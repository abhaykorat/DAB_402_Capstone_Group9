### Dividing data into parts:

### 1) Before Pandemic(Year 2017-2019)
### 2) During Pandemic(Year 2020-2021)

```python
# Selecting data of only year of 2017-2019 and saving it in new variable

Before_Pandemic = data2[(data2.REF_DATE > 2016) & (data2.REF_DATE <= 2019)]

Before_Pandemic
```

```python
# Removing 'Unemployment rate' variable as it contains value in percentage(%) unit.

Before_Pandemic = Before_Pandemic[Before_Pandemic['Labour force characteristics'] != 'Unemployment rate']

Before_Pandemic
```

```python
During_Pandemic = data2[(data2.REF_DATE > 2019) & (data2.REF_DATE <= 2021)]

During_Pandemic
```

```python
During_Pandemic = During_Pandemic[During_Pandemic['Labour force characteristics'] != 'Unemployment rate']

During_Pandemic
```

```python
Before_Pandemic.set_index('REF_DATE', inplace=True)

During_Pandemic.set_index('REF_DATE', inplace=True)
```
