```python
# Loading dataset
Before_Pandemic = pd.read_csv('Before_Pandemic.csv')

During_Pandemic = pd.read_csv('During_Pandemic.csv')
```

### Data Transformation

#### 1. "Before_Pandemic" dataset:

```python
# Renaming features name for better understanding

Before_Pandemic1 = Before_Pandemic.rename(columns={'Labour force characteristics': 'Labour_characteristics', 'North American Industry Classification System (NAICS)':'NAICS'})
Before_Pandemic1 = Before_Pandemic1.replace('Full-time employment','Full-time').replace('Part-time employment','Part-time')

Before_Pandemic1
```

```python
# Using ordinal encoding for categorical variables 'Labour_characteristics', 'Sex', & 'Age group'

Ordinal_encoder = ce.OrdinalEncoder(mapping=[{'col':'Labour_characteristics', 'mapping':{'Unemployment':0, 'Employment':1, 'Full-time':2, 'Labour force':3, 'Part-time':4}},
                                                 {'col':'Sex', 'mapping':{'Both sexes':0, 'Males':1, 'Females':2}},
                                                 {'col':'Age group', 'mapping':{'15 years and over':0, '15 to 24 years':1, '25 to 54 years':2, '55 years and over':3}}])

Ordinal_encoder.fit(Before_Pandemic1)
Before_Pandemic1 = Ordinal_encoder.transform(Before_Pandemic1)

Before_Pandemic1
```

```python
# Using label encoding for categorical variables 'GEO', 'NAICS'


# Changing datatypes of 'GEO' & 'NAICS' variables to category from object
Before_Pandemic1['GEO'] = Before_Pandemic1['GEO'].astype('category')
Before_Pandemic1['NAICS'] = Before_Pandemic1['NAICS'].astype('category')

# Creating an instance of labelencoder
labelencoder = LabelEncoder()


# Assigning numerical values and storing in another column
Before_Pandemic1['GEO_Cat'] = labelencoder.fit_transform(Before_Pandemic1['GEO'])
Before_Pandemic1['NAICS_Cat'] = labelencoder.fit_transform(Before_Pandemic1['NAICS'])

Before_Pandemic1
```

```python
# Creating temporary variable to check assigned numeric values by lable encoding

d1 = Before_Pandemic1[["GEO", "GEO_Cat"]]
d2 = Before_Pandemic1[["NAICS", "NAICS_Cat"]]
```

```python
# Printing assigned numeric values

d1.drop_duplicates('GEO')
d2.drop_duplicates('NAICS')
```

#### 2. "During_Pandemic" dataset:
